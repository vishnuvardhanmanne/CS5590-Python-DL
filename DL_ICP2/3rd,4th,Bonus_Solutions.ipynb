```
%tensorflow_version 1.15
import tensorflow as tf
from keras import Sequential
from keras.datasets import mnist
import numpy as np
from keras.layers import Dense
import matplotlib.pyplot as plt
from keras.utils import to_categorical
```

```
(train_images,train_labels),(test_images, test_labels) = mnist.load_data()
```

```
print(train_images.shape[1:])
```

```
dimData = np.prod(train_images.shape[1:])
print(dimData)
train_data = train_images.reshape(train_images.shape[0],dimData)
test_data = test_images.reshape(test_images.shape[0],dimData)
```

```
train_data = train_data.astype('float')
test_data = test_data.astype('float')
train_data1 = train_data.astype('float')
test_data1 = test_data.astype('float')
```

```
train_data /=255.0
test_data /=255.0
```

```
train_labels_one_hot = to_categorical(train_labels)
test_labels_one_hot = to_categorical(test_labels)
```

```
#3rd Solution:

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(dimData,)))
model.add(Dense(512, activation='tanh'))
model.add(Dense(512, activation='sigmoid'))
model.add(Dense(10, activation='softmax'))
```

```
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
history = model.fit(train_data, train_labels_one_hot, batch_size=256, epochs=10, verbose=1,
                   validation_data=(test_data, test_labels_one_hot))
```

```
#4th Solution:

m1 = Sequential()
m1.add(Dense(512, activation='relu', input_shape=(dimData,)))
m1.add(Dense(512, activation='relu'))
m1.add(Dense(10, activation='softmax'))
m1.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
m1.fit(train_data1, train_labels_one_hot, batch_size=256, epochs=10, verbose=1,
                   validation_data=(test_data1, test_labels_one_hot))
```

```
#Bonus  Solution:

from keras.models import Model
from keras.layers import Input
i=Input(shape=(784,))
x= Dense(512, activation='relu')(i)
x= Dense(512, activation='relu')(x)
x= Dense(10, activation='softmax')(x)
m2 = Model(input=i,output=x)
m2.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
m2.fit(train_data, train_labels_one_hot, batch_size=256, epochs=10, verbose=1,
                   validation_data=(test_data, test_labels_one_hot))
```
